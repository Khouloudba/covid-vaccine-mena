```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

%config InlineBackend.figure_format = 'retina'
```

# Load Covid dataset

```python
df = pd.read_csv('MENAR-CountriesCOVID.csv')
```

```python
df = df.dropna()#### this will drop all rows with a single missing value
df = df.sort_values(by="date")
```

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
missing_values_count
```

# Checking numbers of Countries affected by Covid

```python
df['iso_code'].unique()
```

```python
list_country = ['DZA', 'BHR', 'EGY', 'IRQ', 'ISR', 'JOR', 'KWT', 'LBN', 'LBY',
       'MRT', 'MAR', 'OMN', 'PSE', 'QAT', 'SAU', 'SOM', 'SSD', 'SDN',
       'SYR', 'TUN', 'ARE', 'YEM']

print(len(list_country))### numbers of countries in Mena region
```

```python
new_deaths_ratio = []
new_cases_ratio = []

for i in list_country:
    x = df[df['iso_code']==i]
    new_deaths_rate = sum(x.new_deaths)/len(x)
    new_deaths_ratio.append(new_deaths_rate)
    new_cases_rate = sum(x.new_cases)/len(x)
    new_cases_ratio.append(new_cases_rate)
data = pd.DataFrame({'list_country': list_country,'new_cases_ratio':new_cases_ratio,'new_deaths_ratio':new_deaths_ratio})


new_index = (data['new_cases_ratio'].sort_values(ascending = True)).index.values
sorted_data = data.reindex(new_index)


new_index_1= (data['new_deaths_ratio'].sort_values(ascending = True)).index.values
sorted_data_1 = data.reindex(new_index_1)
```

```python
# visualization
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['list_country'], y=sorted_data['new_cases_ratio'])
plt.xticks(rotation= 45)
plt.xlabel('Mena Countries')
plt.ylabel('new_cases_ratio')
plt.title('new_cases_ratio Given Menar Countries')
```

```python
# visualization
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data_1['list_country'], y=sorted_data_1['new_cases_ratio'])
plt.xticks(rotation= 45)
plt.xlabel('Mena Countries')
plt.ylabel('new_deaths_ratio')
plt.title('new_deaths_ratio Given Menar Countries')
```

# Data analysis of the Spread of Covid over Tunisia

```python
df_1 = df[df['iso_code'] == 'TUN']# Tunisia covid spread data
```

```python
covid_tunisia_total = df_1.groupby(['date'])['total_cases', 'total_deaths'].sum()
covid_tunisia_total=covid_tunisia_total.reset_index().sort_values('date')
covid_tunisia_total.head()
```

```python
df.columns
```

```python
covid_tunisia_new = df_1.groupby(['date'])['new_cases_smoothed', 'new_deaths_smoothed'].max()
covid_tunisia_new=covid_tunisia_new.reset_index().sort_values('date')
covid_tunisia_new.head()
```

```python
#!pip install plotly
```

```python


import plotly.graph_objects as go

x=[i.split('-')[0]+'-'+i.split('-')[1] for i in covid_tunisia_total.date ]
fig = go.Figure()
fig.update_layout(template='plotly_dark')
fig.add_trace(go.Scatter(x=x, 
                         y=covid_tunisia_total['total_cases'],
                         mode='lines+markers',
                         name='total_cases',
                         line=dict(color='rgb(102, 102, 255)', width=2)))
fig.add_trace(go.Scatter(x=x, 
                         y=covid_tunisia_total['total_deaths'],
                         mode='lines+markers',
                         name='total_deaths',
                         line=dict(color='rgb(255, 102, 102)', width=2)))

fig.update_layout(
    title = 'Spread of COVID-19 over time in Tunisia per total cases and total deaths',
    xaxis_tickformat = '%d %B (%a)<br>%Y'
)
fig.show()
```

```python
import plotly.graph_objects as go

x=[i.split('-')[0]+'-'+i.split('-')[1] for i in covid_tunisia_new.date ]
fig = go.Figure()
fig.update_layout(template='plotly_dark')
fig.add_trace(go.Scatter(x=x, 
                         y=covid_tunisia_new['new_cases_smoothed'],
                         mode='lines+markers',
                         name='new_cases_smoothed',
                         line=dict(color='rgb(102, 102, 255)', width=2)))
fig.add_trace(go.Scatter(x=x, 
                         y=covid_tunisia_new['new_deaths_smoothed'],
                         mode='lines+markers',
                         name='new_deaths_smoothed',
                         line=dict(color='rgb(255, 102, 102)', width=2)))

fig.update_layout(
    title = 'Spread of COVID-19 over time in Tunisia per new cases and new deaths smoothed',
    xaxis_tickformat = '%d %B (%a)<br>%Y'
)
fig.show()
```
